# PyFraug

```python
import time

import pyfraug as pf
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../../data/Car/Car.csv").to_numpy()

x = data[:,:-1].astype(np.float64)
y = list(map(lambda a: str(a), data[:,-1]))
```

```python
dataset = pf.Dataset(x, y)
```

```python
# pipeline = (pf.AugmentationPipeline()
#             + pf.Repeat(10)
#             + pf.Crop(100)
#             + pf.ConditionalAugmenter(
#                 pf.AddNoise(pf.NoiseType.Slope, bounds=(0.01, 0.02)),
#                 0.5
#             )
#             + pf.Jittering(0.1))
aug = pf.AugmentationPipeline() + pf.AddNoise(pf.NoiseType.Gaussian, mean=0, std_dev=0.1)
```

```python
start = time.time()
aug.augment_dataset(dataset, parallel=True)
end = time.time()
print(f"Augmentation took {end - start:.4f} seconds")
```
